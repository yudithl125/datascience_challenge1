### Importación de datos



```python
import matplotlib.pyplot as plt
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
tiendas = ['tienda1', 'tienda2', 'tienda3', 'tienda4']
total_ventas = []
total_ventas.append(tienda['Precio'].sum())
total_ventas.append(tienda2['Precio'].sum())
total_ventas.append(tienda3.Precio.sum())
total_ventas.append(tienda4.Precio.sum())
def mostrar_valor_y_porcentaje(pct, all_vals):
    valor_absoluto = int(round(pct/100.*sum(all_vals)))
    return f'{valor_absoluto} \n({pct:.1f}%)'
plt.figure(figsize=(6, 6))
plt.pie(total_ventas, labels=tiendas, autopct=lambda pct: mostrar_valor_y_porcentaje(pct, total_ventas), startangle=90)
plt.title('Distribución de ventas por tienda')
plt.axis('equal')
plt.show()
```

# 2. Ventas por categoría

```python
plt.rcParams.update({
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
})

print(tienda['Categoría del Producto'].value_counts())
plt.bar(tienda['Categoría del Producto'].value_counts().index, tienda['Categoría del Producto'].value_counts())
plt.xticks(rotation=90)
plt.show()
print(tienda2['Categoría del Producto'].value_counts())
plt.bar(tienda['Categoría del Producto'].value_counts().index, tienda['Categoría del Producto'].value_counts())
plt.xticks(rotation=90)
plt.show()
print(tienda3['Categoría del Producto'].value_counts())
plt.bar(tienda['Categoría del Producto'].value_counts().index, tienda['Categoría del Producto'].value_counts())
plt.xticks(rotation=90)
plt.show()
print(tienda4['Categoría del Producto'].value_counts())
plt.bar(tienda['Categoría del Producto'].value_counts().index, tienda['Categoría del Producto'].value_counts())
plt.xticks(rotation=90)
plt.show()
```

# 3. Calificación promedio de la tienda


```python
tiendas = ['tienda1', 'tienda2', 'tienda3', 'tienda4']
promedio_calificacion = []
suma1 = tienda['Calificación'].sum()
suma2 = tienda2['Calificación'].sum()
suma3 = tienda3['Calificación'].sum()
suma4 = tienda4['Calificación'].sum()

longitud1 = tienda['Calificación'].size
longitud2 = tienda2['Calificación'].size
longitud3 = tienda3['Calificación'].size
longitud4 = tienda4['Calificación'].size

promedio_calificacion.append(suma1/longitud1)
promedio_calificacion.append(suma2/longitud2)
promedio_calificacion.append(suma3/longitud3)
promedio_calificacion.append(suma4/longitud4)

plt.rcParams.update({
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
})
print(promedio_calificacion)
plt.barh(tiendas, promedio_calificacion)

for i, valor in enumerate(promedio_calificacion):
    plt.text(valor - 1, i, str(round(valor,3)), va='center', ha='right', color='white', fontweight='bold')

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
