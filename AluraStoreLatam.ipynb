### Importación de datos



```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.patches import Patch

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
tiendas = ['tienda1', 'tienda2', 'tienda3', 'tienda4']
total_ventas = []
total_ventas.append(tienda['Precio'].sum())
total_ventas.append(tienda2['Precio'].sum())
total_ventas.append(tienda3.Precio.sum())
total_ventas.append(tienda4.Precio.sum())
def mostrar_valor_y_porcentaje(pct, all_vals):
    valor_absoluto = int(round(pct/100.*sum(all_vals)))
    return f'{valor_absoluto} \n({pct:.1f}%)'
plt.figure(figsize=(6, 6))
plt.pie(total_ventas, labels=tiendas, autopct=lambda pct: mostrar_valor_y_porcentaje(pct, total_ventas), startangle=90)
plt.title('Distribución de ventas por tienda')
plt.axis('equal')
plt.show()
```

# 3. Calificación promedio de la tienda


```python
tiendas = ['tienda1', 'tienda2', 'tienda3', 'tienda4']
promedio_calificacion = []
suma1 = tienda['Calificación'].sum()
suma2 = tienda2['Calificación'].sum()
suma3 = tienda3['Calificación'].sum()
suma4 = tienda4['Calificación'].sum()

longitud1 = tienda['Calificación'].size
longitud2 = tienda2['Calificación'].size
longitud3 = tienda3['Calificación'].size
longitud4 = tienda4['Calificación'].size

promedio_calificacion.append(suma1/longitud1)
promedio_calificacion.append(suma2/longitud2)
promedio_calificacion.append(suma3/longitud3)
promedio_calificacion.append(suma4/longitud4)

plt.rcParams.update({
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
})

plt.barh(tiendas, promedio_calificacion)

for i, valor in enumerate(promedio_calificacion):
    plt.text(valor - 1, i, str(round(valor,3)), va='center', ha='right', color='white', fontweight='bold')

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
def top_ventas_por_producto(tienda_actual):
  prod_venta = {}
  for producto_actual in tienda_actual['Producto']:
      #El Producto esta en el diccionario? Si no es asi lo agrega al diccionario
      #Caso1: El Producto no esta en el diccionario, lo agregamos con una venta
      if producto_actual not in prod_venta:
        prod_venta[producto_actual] = 1
        #Caso2: El producto ya esta en el diccionario, lo buscamos e incrementamos en 1
      else:
        prod_venta[producto_actual] = prod_venta[producto_actual]+1
      #Cuantas veces el producto se vendio?
  top_3_mas_vendidos = dict(sorted(prod_venta.items(), key=lambda item: item[1], reverse=True)[:3])
  top_3_menos_vendidos = dict(sorted(prod_venta.items(), key=lambda item: item[1], reverse=False)[:3])
  return top_3_mas_vendidos | top_3_menos_vendidos

def grafico_ventas(diccionario,texto):
  colores = ['skyblue'] * 3 + ['lightcoral'] * 3
  leyenda = [
    Patch(color='skyblue', label='MAX'),
    Patch(color='lightcoral', label='MIN')
  ]
  plt.bar(diccionario.keys(), diccionario.values(), color=colores)
  plt.xlabel('Producto')
  plt.ylabel('Ventas')
  plt.title('Top Ventas por Producto de la '+ texto)
  plt.legend(handles=leyenda)
  plt.grid(axis='y', linestyle='--', alpha=0.7)
  plt.xticks(rotation=90)
  plt.show()

tops_de_la_tienda1 = top_ventas_por_producto(tienda)
grafico_ventas(tops_de_la_tienda1,'Tienda N°1')
print()
tops_de_la_tienda2 = top_ventas_por_producto(tienda2)
grafico_ventas(tops_de_la_tienda2,'Tienda N°2')
print()
tops_de_la_tienda3 = top_ventas_por_producto(tienda3)
grafico_ventas(tops_de_la_tienda3,'Tienda N°3')
print()
tops_de_la_tienda4 = top_ventas_por_producto(tienda4)
grafico_ventas(tops_de_la_tienda4,'Tienda N°4')
print()
```

# 5. Envío promedio por tienda

```python
tiendas = ['tienda1', 'tienda2', 'tienda3', 'tienda4']
promedio_costo_de_envios = []
suma1 = tienda['Costo de envío'].sum()
suma2 = tienda2['Costo de envío'].sum()
suma3 = tienda3['Costo de envío'].sum()
suma4 = tienda4['Costo de envío'].sum()

longitud1 = tienda['Costo de envío'].size
longitud2 = tienda2['Costo de envío'].size
longitud3 = tienda3['Costo de envío'].size
longitud4 = tienda4['Costo de envío'].size

promedio_costo_de_envios.append(suma1/longitud1)
promedio_costo_de_envios.append(suma2/longitud2)
promedio_costo_de_envios.append(suma3/longitud3)
promedio_costo_de_envios.append(suma4/longitud4)

plt.rcParams.update({
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
})
plt.barh(tiendas, promedio_costo_de_envios)
for i, valor in enumerate(promedio_costo_de_envios):
    plt.text(valor - 1, i, str(round(valor,2)), va='center', ha='right', color='white', fontweight='bold')

plt.xticks(rotation=90)
plt.tight_layout()
plt.title('Promedio de Costo de Envío por Tienda')
plt.show()
```

# 6. Desempeño Geográfico

```python
# Redondeamos coordenadas para agrupar por zonas aproximadas
tienda3['lat_r'] = tienda3['lat'].round(2)
tienda3['lon_r'] = tienda3['lon'].round(2)
tienda3['Eficiencia'] = 1 - (tienda3['Costo de envío'] / tienda3['Precio'])

# Agrupamos por zona geográfica
agrupado = tienda3.groupby(['lat_r', 'lon_r']).agg({
    'Eficiencia': 'mean'
}).reset_index()

# Graficamos
plt.figure(figsize=(10, 8))
scatter = plt.scatter(
    agrupado['lon_r'], agrupado['lat_r'],
    c=agrupado['Eficiencia'],
    cmap='RdYlGn',  # Rojo = ineficiente, Verde = eficiente
    s=100, edgecolor='k'
)
plt.colorbar(scatter, label='Eficiencia (1 - Costo / Precio)')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.title('Eficiencia de envío por zona geográfica (Tienda 3)')
plt.grid(True)
plt.gca().set_facecolor('white')
plt.show()
```

```python
# Función para preparar los datos
def preparar_datos(df, nombre_tienda):
    df = df.copy()
    df['lat_r'] = df['lat'].round(2)
    df['lon_r'] = df['lon'].round(2)
    df['Eficiencia'] = 1 - (df['Costo de envío'] / df['Precio'])
    df_grouped = df.groupby(['lat_r', 'lon_r']).agg({'Eficiencia': 'mean'}).reset_index()
    df_grouped['Tienda'] = nombre_tienda
    return df_grouped

# Preparar cada tienda
datos1 = preparar_datos(tienda, 'Tienda 1')
datos2 = preparar_datos(tienda2, 'Tienda 2')
datos3 = preparar_datos(tienda3, 'Tienda 3')
datos4 = preparar_datos(tienda4, 'Tienda 4')

# Lista de datos
todas = [datos1, datos2, datos3, datos4]
titulos = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Crear subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
axes = axes.flatten()

for i, datos in enumerate(todas):
    sc = axes[i].scatter(
        datos['lon_r'], datos['lat_r'],
        c=datos['Eficiencia'],
        cmap='RdYlGn',
        s=100, edgecolor='black'
    )
    axes[i].set_title(titulos[i])
    axes[i].set_xlabel('Longitud')
    axes[i].set_ylabel('Latitud')
    axes[i].grid(True)
    axes[i].set_facecolor('white')

# Barra de color común
fig.colorbar(sc, ax=axes, orientation='horizontal', fraction=0.05, pad=0.05, label='Eficiencia (1 - Costo / Precio)')
plt.suptitle('Mapa de calor de eficiencia logística por tienda', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
